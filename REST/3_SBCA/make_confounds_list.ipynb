```python
# create nuisance regressors

# Ikko Kimura, Osaka University Hospital, 2021/8/24 
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
WD="/project/amano-g/kimura/PDRegistry/"
subj_list=np.loadtxt("subj_list_all.txt",dtype="str")

for i in range(len(subj_list)):
    #subj="93"
    subj=subj_list[i]
    print(subj)
    if int(subj)<100:
        nuisance=pd.read_table(WD+"output/fmriprep/sub-"+subj+"/func/sub-"+subj+"_task-rest_desc-confounds_timeseries.tsv")
    else:
        nuisance=pd.read_table(WD+"output/fmriprep/sub-"+subj+"/func/sub-"+subj+"_task-rest_desc-confounds_timeseries.tsv")

    #print(nuisance)
    nuisance_bk=nuisance
    nuisance=nuisance[4::]
    unc_num=nuisance[(nuisance["framewise_displacement"]<0.5) & (nuisance["std_dvars"]<1.5)].shape
    df=pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)

    for fname in ["csf","white_matter","global_signal","trans_x","trans_y","trans_z","rot_x","rot_y","rot_z"]:
        df[fname]=nuisance_bk[fname]

    for i in range(236-unc_num[0]):
        if i<10:
            df["motion_outlier0"+str(i)]=nuisance_bk["motion_outlier0"+str(i)]
        else:
            df["motion_outlier"+str(i)]=nuisance_bk["motion_outlier"+str(i)]

    if int(subj)<100:
        df.to_csv(WD+"output/fmriprep/sub-"+subj+"/func/sub-"+subj+"_task-rest_desc-confounds_timeseries_post2.tsv", sep='\t', index=False)
    else:
        df.to_csv(WD+"output/fmriprep/sub-"+subj+"/func/sub-"+subj+"_task-rest_desc-confounds_timeseries_post2.tsv", sep='\t', index=False)
```
